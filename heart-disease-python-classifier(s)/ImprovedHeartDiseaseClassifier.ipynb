```python
import numpy as np
import pandas as pd
import sklearn

url = "https://raw.githubusercontent.com/Sanket2311/HeartDisease-classification-using-knn-LogisticRegression-and-SVM-/master/heart.csv"
data = pd.read_csv(url)

print(data)
```

```python
from sklearn.utils import shuffle
data = shuffle(data)

X = data.iloc[:, 0:13]
y = data.iloc[:, 13:]

from sklearn import preprocessing
le = preprocessing.LabelEncoder()
y = y.apply(le.fit_transform)

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.50)

print(len(X_train))
```
