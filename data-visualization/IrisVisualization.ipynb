```python
import seaborn as sns

sns.set(style="ticks")

df=sns.load_dataset("iris")

sns.pairplot(df, hue="species")
```

```python
print(df)
```

```python
data_csv = df.to_csv()

import numpy as np

data_array = np.array(df.values)


print(data_array)
```

```python
petal_lengths = data_array[:, 2]
petal_widths = data_array[:, 3]

print(petal_lengths)
print(petal_widths)
```

```python
import matplotlib.pyplot as plt

plt.plot(petal_widths, petal_lengths)
plt.xlabel("petal widths")
plt.ylabel("petal lengths")
plt.show()
```
